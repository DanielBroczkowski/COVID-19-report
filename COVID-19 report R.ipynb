##### Mapa świata przedstawiająca ilość zgonów z powodu zarażenia koronawirusem w poszczególnych państwach

```r
#install.packages(c("cowplot", "googleway", "ggplot2", "ggrepel","ggspatial", 
#"libwgeom", "sf", "rnaturalearth", "rnaturalearthdata", "rgeos", "maps", "viridis"))
options(repr.matrix.max.rows=270, repr.matrix.max.cols=40)

library("viridis")
library("ggplot2")
library("sf")
library("rnaturalearth")
library("rnaturalearthdata")
library("rgeos")
require("maps")

theme_set(theme_bw())
world <- ne_countries(scale = "medium", returnclass = "sf")

deaths = Deaths_raw[,ncol(Deaths_raw)]
deaths <- as.matrix(deaths)
colnames(deaths) <- "deaths"

death_countries = Deaths_raw[,2]
death_countries <- as.matrix(death_countries)
colnames(death_countries) <- "region"

death_note <- data.frame(
    region = death_countries,
    casualities = deaths
)

death_note = death_note %>% 
 group_by(region) %>% 
 summarise_all(funs(sum))

death_note[death_note == "US"] <- "USA"
death_note[death_note == "United Kingdom"] <- "UK"

my_map <- map_data("world")
Death_map <- left_join(my_map, death_note, by = "region")

ggplot(Death_map, aes(long, lat, group = group)) +
  geom_polygon(aes(fill = deaths), color = "white") + scale_fill_viridis_c(option = "plasma", trans = "identity")
```
