```python
import pandas as pd
import numpy as np


MultipleData = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/who_covid_19_situation_reports/who_covid_19_sit_rep_time_series/who_covid_19_sit_rep_time_series.csv")
pd.set_option("display.max_rows", 250)
pd.set_option("display.max_columns", 250)
MultipleData
```

```python
ConfirmedCases_raw = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
pd.set_option("display.max_rows", 600)
pd.set_option("display.max_columns", 250)
ConfirmedCases_raw
```

```python
Deaths_raw = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
pd.set_option("display.max_rows", 600)
pd.set_option("display.max_columns", 250)
Deaths_raw
```

```python
Recoveries_raw = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
pd.set_option("display.max_rows", 600)
pd.set_option("display.max_columns", 250)
Recoveries_raw
```

```python
MultipleData_Plot = MultipleData.filter(regex='\d{1,2}\/\d{1,2}\/\d{4}', axis=1).filter(regex='^0$', axis=0).rename({0: 'cases'}, axis=0)
MultipleData_Plot = MultipleData_Plot.astype(float)
MultipleData_Plot = MultipleData_Plot.T.plot.bar(figsize=(20,8), title='An increase in cases in the first month of the pandemic')
MultipleData_Plot = MultipleData_Plot.get_legend().remove()
```

```python
UE=['Poland', 'France', 'Germany']
ConfirmedCasesEurope_raw = ConfirmedCases_raw[ConfirmedCases_raw['Country/Region'].isin(UE)]
ConfirmedCasesEurope_raw = ConfirmedCasesEurope_raw.groupby('Country/Region').sum()
ConfirmedCasesEurope_raw = ConfirmedCasesEurope_raw.filter(regex='\d{1,2}\/\d{1,2}\/\d{2}', axis=1)
ConfirmedCasesEurope_raw.T.plot(figsize=(20,8))
```

```python
ConfirmedCasesAmerica_raw = ConfirmedCases_raw[ConfirmedCases_raw['Country/Region'] == 'US']
print(ConfirmedCases_raw)
ConfirmedCasesAmerica_raw = ConfirmedCasesAmerica_raw.set_index('Province/State', drop=True).filter(regex='\d{1,2}\/\d{1,2}\/\d{2}', axis=1).filter(regex='^[a-zA-Z0-9_ ]*$', axis=0)
ConfirmedCasesAmerica_raw.T.plot(figsize=(20,8))
```

```python
import matplotlib.pyplot as plt

ConfirmedPl = ConfirmedCases_raw[ConfirmedCases_raw['Country/Region']=='Germany']
ConfirmedPl = ConfirmedPl.set_index('Country/Region', drop=True).filter(regex='\d{1,2}\/\d{1,2}\/\d{2}', axis=1)

LatestConfirmedPoland = ConfirmedPl.iloc[:,-1].Germany
print(LatestConfirmedPoland)

DeathsPoland = Deaths_raw[Deaths_raw['Country/Region']=='Germany']
DeathsPoland = DeathsPoland.set_index('Country/Region', drop=True).filter(regex='\d{1,2}\/\d{1,2}\/\d{2}', axis=1)

LatestDeathPoland = DeathsPoland.iloc[:,-1].Germany
print(LatestDeathPoland)

RecoveriesPoland = Recoveries_raw[Recoveries_raw['Country/Region']=='Germany']
RecoveriesPoland = RecoveriesPoland.set_index('Country/Region', drop=True).filter(regex='\d{1,2}\/\d{1,2}\/\d{2}', axis=1)

LatestRecoveriesPoland = RecoveriesPoland.iloc[:,-1].Germany
print(LatestRecoveriesPoland)

labels = ['Confirmed', 'Deaths', "Recoveries"]
sizes = [LatestConfirmedPoland, LatestDeathPoland, LatestRecoveriesPoland]
colors = ['cyan', 'red', 'green']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, wedgeprops=dict(width=0.3), startangle=140)
plt.legend(patches, sizes + labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.title("Germany")
plt.show()

#ConfirmedPl = ConfirmedPl.T.plot(kind='pie', figsize=(20,8), subplots=True)

###################################################################################################

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))


recipe = [str(LatestConfirmedPoland) + " confirmed",
          str(LatestDeathPoland) + " deaths",
          str(LatestRecoveriesPoland) + " recovered"]

data = [LatestConfirmedPoland, LatestDeathPoland, LatestRecoveriesPoland]

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-90)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Germany")

plt.show()
```
